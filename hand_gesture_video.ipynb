```python
import cv2
import os

def parseSingeVideo(input_dir, output_file):
  
  # Get the list of images in the input directory
  image_list = os.listdir(input_dir)
  sorted_image_list = sorted(image_list)  # Sort the image list if needed

  # Set the frame size and frame rate
  frame_width, frame_height = cv2.imread(os.path.join(input_dir, sorted_image_list[0])).shape[:2]
  fps = 30

  # Create a video writer object
  fourcc = cv2.VideoWriter_fourcc(*"mp4v")
  out = cv2.VideoWriter(output_file, fourcc, fps, (frame_height, frame_width))

  # Iterate through the sorted image list and add each image to the video
  for image_name in sorted_image_list:
      if image_name != "output.avi":
        image_path = os.path.join(input_dir, image_name)
        image = cv2.imread(image_path)
        out.write(image)
  
  # Release the video writer object and destroy any remaining windows
  out.release()
  cv2.destroyAllWindows()
```

```python
# Parse Visualization Video
vis_dir = "/content/gdrive/MyDrive/Hand_Research/TimeSformer-main/output/vit-explain"
vis_output = "/content/gdrive/MyDrive/Hand_Research/TimeSformer-main/output/vit-explain/output.mp4"
parseSingeVideo(vis_dir, vis_output)
```
